# Sound Intensity
The intensity of a sound wave is defined as
\begin{equation}
I=\frac{1}{T}\int_0^Tp(t)u(t)dt
\end{equation}
the relation between particle velocity $u(t)$ and pressure gradient is given by
\begin{equation}
\frac{\partial p}{\partial x}=-\rho_o\frac{\partial u}{\partial t}
\end{equation}
which expresses the observation that particle have a mass that will resist a change in speed under the influence of an applied force (inertial reaction).

The paricle velocity is obtained by integration and using the Euler approximation to the pressure gradient one gets
\begin{equation}
u(t)=u(0)-\frac{1}{\rho \Delta x}\int_0^t{\Delta p(\tau) d\tau}
\end{equation}

Usin $u(0)=0$ one ontains fot the sound intensity
\begin{equation}
I=-\frac{1}{T}\frac{1}{\rho \Delta x}\int_0^Tp(t)\Biggl(\int_0^t{\Delta p(\tau) d\tau}\Biggr)dt
\end{equation}

Using next the spectral equivalent
\begin{equation}
p(t)=\frac{1}{2\pi}\int_{-\pi}^{\pi}P(\omega)\exp(-i\omega t)d\omega
\end{equation}
and with

\begin{equation}
\int_{0}^{t}\exp(-i\omega \tau)d\tau=\frac{i}{\omega}\Biggl(\exp(-i\omega t)-1\Biggr)
\end{equation}

\begin{equation}
\int_0^t \Delta p(\tau)d\tau=
      \frac{i}{2\pi}\int_{-\pi}^{\pi}\Biggl(\frac{\Delta P(\omega)}{\omega}\Biggr)\Biggl(\exp(-i\omega t)-1\Biggr)d\omega =
      \frac{i}{2\pi}\int_{-\pi}^{\pi}\Biggl(\frac{\Delta P(\omega)}{\omega}\Biggr)\exp(-i\omega t)d\omega
\end{equation}

we obtain
\begin{equation}
I=\frac{1}{T}\int_0^Tp(t)u(t)dt 
\end{equation}

\begin{equation}
I=-\frac{1}{T}\frac{1}{\rho \Delta x}\int_0^T p(t)\Biggl(\int_0^t{\Delta p(\tau) d\tau}\Biggr)dt
\end{equation}

\begin{equation}
I=-\frac{i}{\rho \Delta x}\frac{1}{2\pi}\frac{1}{2\pi} 
\int_{-\pi}^{\pi} \int_{-\pi}^{\pi} P(\omega') 
\Biggl(\frac{\Delta P(\omega)}{\omega}\Biggr) \Biggl(\frac{1}{T}\int_0^T  \exp(-i\omega' t)\exp(-i\omega t) dt \Biggr) d\omega d\omega' 
\end{equation}

Assume the temporal ingration
\begin{equation}
\frac{1}{T}\int_0^T  \exp(-i\omega' t)\exp(-i\omega t) dt = \delta (\omega' + \omega = 0)
\end{equation}
evaluates to one if and only if $\omega'+\omega=0$ and to zero otherwise, so that

\begin{equation}
I=-\frac{i}{\rho \Delta x}\frac{1}{2\pi}\frac{1}{2\pi} 
\int_{-\pi}^{\pi} \int_{-\pi}^{\pi} P(-\omega) 
\Biggl(\frac{\Delta P(\omega)}{\omega}\Biggr)  d\omega d\omega' 
\end{equation}
or
\begin{equation}
I=-\frac{i}{\rho \Delta x}\frac{1}{2\pi}
\int_{-\pi}^{\pi} P^*(\omega) \Biggl(\frac{\Delta P(\omega)}{\omega}\Biggr)  d\omega
\end{equation}

Approximating the sound pressure and pressure gradient by pressure measurements of 2 sensors
\begin{equation}
P(\omega)=\frac{P_2(\omega)+P_1(\omega)}{2}
\end{equation}

\begin{equation}
\Delta P(\omega)=P_2(\omega)-P_1(\omega)
\end{equation}

\begin{equation}
P^*(\omega)\Delta P(\omega) = \frac{1}{2}\{(P_2^*(\omega)+P_1^*(\omega)) (P_2(\omega)-P_1(\omega))\}
\end{equation}

\begin{equation}
P^*(\omega)\Delta P(\omega) = \frac{1}{2}\{ |P_2(\omega)|^2 - |P_1(\omega)|^2 +P_2(\omega)P_1^*(\omega)-P_2^*(\omega)P_1(\omega)\}
\end{equation}

With $|P_2(\omega)|^2 = |P_1(\omega)|^2$
\begin{equation}
P^*(\omega)\Delta P(\omega) = \frac{1}{2}\{ P_2(\omega)P_1^*(\omega)-P_2^*(\omega)P_1(\omega)\}
\end{equation}
or
\begin{equation}
P^*(\omega)\Delta P(\omega) = i\text{Im}\{ P_2(\omega)P_1^*(\omega)\}
\end{equation}
and finally the spectral sound intesity is approcimated by
\begin{equation}
I(f)=-\frac{i}{\rho \Delta x} \frac{ P^*(\omega)\Delta P(\omega)}{\omega}
    =-\frac{1}{\rho \Delta x} \frac{\text{Im}\{P_2(\omega) P_1^* (\omega)\}}{\omega}
\end{equation}
